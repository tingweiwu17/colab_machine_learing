```python
import matplotlib.pyplot as plt
import matplotlib.style
import pandas as pd
import numpy as np

matplotlib.style.use('ggplot')
```

```python
df = pd.read_csv('http://bit.ly/kaggletrain')
```

```python
df
```

```python
fig, ax = plt.subplots()

a = np.array( [df["Sex"] == "male"] ) #判斷是否是male=>顯示True/False

male = np.sum(a > 0) #True的數量=male數量

total = a.shape[1]  #總共的數量

female = total - male #female數量

x = [1, 2]
y = [male, female]
labels = ['male', 'female']

ax.bar(x, y, tick_label=labels, label='y')

plt.show()
```

```python
fig, ax = plt.subplots()

df_selected1 = df[df["Pclass"] == 1] #Pclass1的客人數量
df_selected1.shape[0]

df_selected1_1 = df_selected1[df_selected1["Survived"] == 1] #Pclass1且存活數量

df_selected1_0 = df_selected1.shape[0] - df_selected1_1.shape[0] #Pclass1且死亡數量


df_selected2 = df[df["Pclass"] == 2] #Pclass2的客人數量
df_selected2.shape[0]

df_selected2_1 = df_selected2[df_selected2["Survived"] == 1] #Pclass2且存活數量

df_selected2_0 = df_selected2.shape[0] - df_selected2_1.shape[0] #Pclass2且死亡數量


df_selected3 = df[df["Pclass"] == 3] #Pclass3的客人數量
df_selected3.shape[0]

df_selected3_1 = df_selected3[df_selected3["Survived"] == 1] #Pclass3且存活數量

df_selected3_0 = df_selected3.shape[0] - df_selected3_1.shape[0] #Pclass3且死亡數量



# 取得相應數值放數長條圖當中
x = [1, 2, 3]
y1 = [df_selected1_0, df_selected2_0, df_selected3_0]
y2 = [df_selected1_1.shape[0], df_selected2_1.shape[0], df_selected3_1.shape[0]]
labels = ['Pclass1', 'Pclass2', 'Pclass3']

width = 0.4  #設定長條圖寬度
ax.bar(x, y1, width=width, tick_label=labels,
       label='Dead') #繪製長條圖_Dead部分

x2 = [num + width for num in x]
ax.bar(x2, y2, width=width, label='Live') #繪製長條圖_Live部分

ax.legend()

plt.show()
```

```python
fig, ax = plt.subplots()

ax.hist(df["Age"], bins=20)  # 取得數值與顯示直方圖數量

plt.show()
```

```python
fig, ax = plt.subplots()

df_live = df[df["Survived"] ==0]
x = df_live["Age"]
y = df_live["Fare"]

ax.scatter(x, y, marker='x')  # 繪製散佈圖

df_live = df[df["Survived"] ==1]
x = df_live["Age"]
y = df_live["Fare"]

ax.scatter(x, y)  # 繪製散佈圖

plt.show()
```
